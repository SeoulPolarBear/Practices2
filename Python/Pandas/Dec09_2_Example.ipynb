```python
#필요한 라이브러리 불러오기
from bs4 import BeautifulSoup
import urllib.request as req
import requests
import pandas as pd
import html5lib
```

```python
#https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss212

# 전체 내용 데이터 파싱하기
# parser 종류 : html.parser -> html5lib

url = "https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss212"

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html5lib') # html.parser / html5lib / lxml / ...
print(soup)
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용을 출력

# select() vs select_one()
# select() : 매칭되는 여러 값들을 리스트에 담음...
# select_one() : 매칭되는 값 하나만 가져와서 => 객체

# select관련 => 하나의 열에 여러 선택자가 필요한 경우
#     => .000 or .xxx => (or : ','(쉼표))

# 속성 값 구해야 할 때 : []
rankTop = soup.select('.numTop')
rankNum = soup.select('.rankNum')
#print(type(ranks)) # list
ranks = rankTop + rankNum
print(ranks)

restName = soup.select(".restName")
print(restName)

listType = soup.select(".listType")
print(listType)

listArea = soup.select(".listArea")
print(listArea)

url = soup.select(".restName > a")
print(url) 
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용을 출력
for i in range(len(ranks)):
  ranks[i] = ranks[i].get_text()

for i in range(len(restName)):
  restName[i] = restName[i].get_text()

for i in range(len(url)):
  url[i] = url[i].get_attribute_list('href')[0]
  url[i] = f"https://www.menupan.com{url[i]}"
  
for i in range(len(listType)):
  listType[i] = listType[i].get_text()

for i in range(len(listArea)):
  listArea[i] = listArea[i].get_text()

```

```python
print(ranks)
print(restName)
print(listType)
print(listArea)
print(url)
```

```python
dict = {'ranks': ranks,
        'restName': restName,
        'listType': listType,
        'listArea': listArea,
        'url': url,
        }
print(dict)
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용을 출력

# select() vs select_one()
# select() : 매칭되는 여러 값들을 리스트에 담음...
# select_one() : 매칭되는 값 하나만 가져와서 => 객체

# select관련 => 하나의 열에 여러 선택자가 필요한 경우
#     => .000 or .xxx => (or : ','(쉼표))

# 속성 값 구해야 할 때 : []
#body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul
s = "body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul"
ulTag = soup.select_one(s)

for li in ulTag.select("li"):
  print('순위 : ' ,li.select_one('.numTop, .rankNum').string)
  print('상호 : ' ,li.select_one('.restName').string)
  print('업종 : ' ,li.select_one('.listType').string)
  print('주소 : ' ,li.select_one('.listArea').string)
  #print('url : ' ,li.select_one('.restName a')['href'])
  print('url : ' ,li.select_one('.listPhoto a')['href'])
  # img의 경로지정 속성은 src(그 경로에 있는 이미지 불러오겠다.)
  # a태그의 url 경로지정 속성은 href(그 경로로 이동하겠다.)
```

```python
# 위의 데이터들을 list로 나타내는 작업
# [{'순위' : '1', '상호' : 'xxxx', '업종': '한식',...},
#  ...]

# url에 대한 내용은 따로 별개의 list에 넣기
dic_menupan=[]
list_url=[]
for li in ulTag.select("li"):
  dic_menupan.append({'순위': int(li.select_one('.numTop, .rankNum').string),
                  '상호' : li.select_one('.restName').string,
                  '업종' : li.select_one('.listType').string,
                  '주소' : li.select_one('.listArea').string})
  list_url.append('https://www.menupan.com' + li.select_one('.listPhoto a')['href'])
print(dic_menupan)
print(list_url)
```

```python
# menuList를 가지고 순위를 index로 하는 DataFrame을 생성
menu = pd.DataFrame({},columns=['상호','업종','주소'])
print(menu)
for m in dic_menupan:
#   menu[m['순위']] = m
  menu.loc[m['순위']] = [m['상호'],m['업종'],m['주소']]
print(menu)
print(menu.index)
menu

#============================================================================================
menuDF = pd.DataFrame(dic_menupan)
menuDF.set_index('순위', inplace=True)
menuDF
```

```python
# url로 들어가서 전화번호와 상세주소 가져온 후에 
# 만들어놓은 DataFrame에 붙이기 ( + '주소'열은 삭제)

phoneNumber = []
addr =[]

pn ="body > center > div.WrapMain > div.mainArea01 > div.areaBasic > dl.restTel > dd"
ad ="body > center > div.WrapMain > div.mainArea01 > div.areaBasic > dl.restAdd > dd.add1 > a"

for i in range(len(list_url)):
  response2 = requests.get(list_url[i])
  soup2 = BeautifulSoup(response2.content, 'html5lib')
  phoneNumber.append(soup2.select_one(pn).string)
  addr.append(soup2.select_one(ad).string)

print(phoneNumber)
print(addr)
```

```python
menuDF['전화번호'] = phoneNumber
menuDF['상세주소'] = addr
menuDF.drop(['주소'], axis=1, inplace=True)
menuDF
```
