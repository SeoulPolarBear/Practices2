# 함수 매핑(mapping)
시리즈나 데이터프레임의 개별 요소를 특정 함수에 1:1로 대응시키는 과정

사용자가 직접 만든 함수(람다 포함)를 적용시킬 수 있기 때문에

for 문이나 기본 함수로 처리하는 것보다 훨씬 효율적인 장점

```python
#seaborn에서 타이타닉 데이터 불러와서 'age', 'fare'열만 인덱싱
t = sns.load_dataset('titanic')[['age','fare']]
#     => 'ten' 열을 만들어서 => 10을 넣기

t['ten'] = 10
t
```

```python
# 값을 넣으면 10을 더해주는 함수
def sum10(num):
  return num + 10
# 값을 2개 넣으면 그 둘을 곱해주는 함수
def multi(a, b):
  return a * b
# 값을 3개 넣으면 그 셋을 더해주는 함수
def sum_Three(a,b,c):
  return a + b + c
```

```python
# 이 함수들을 apply()함수들을 이용해서 시리즈객체의 개별요소에 적용
sr1 = t['age'].apply(sum10)
sr1.head()
```

```python
sr2 = t['fare'].apply(multi, b=0.9) # a에는 t['fare'] 
# 첫번째 파라미터 : 함수
# 두번째 파라미터 :값
sr2.head()
```

```python
sr3 = t['fare'].apply(sum_Three, b=10.1, c = 20) 
sr3.head()
```

```python
# 'ten'열의 값들을 lambda함수를 이용해서 sum10함수 적용
sr4 = t['ten'].apply(lambda a : sum10(a))
sr4.head()
```

# apply() vs applymap()

```python
df_map = t.applymap(sum10) # 데이터프레임의 모든 요소에 함수를 적용함
df_map
```

# 시리즈 객체에 함수 매핑

DataFrame객체.apply(매핑함수, axis=값)
열 또는 행에 대해서 함수를 적용시킬 수 있음

```python
# 뭔가를 넣었을 때  결측값을 확인하는 함수
def isNaN(series):
  return series.isnull()
```

```python
# 그 함수를 데이터프레임의 열에 매핑
result = df_map.apply(isNaN,axis=1)
print(result)
result.describe()
```

# 3. 데이터프레임 객체에 함수를 매핑

DataFrame객체.pipe() 함수
pipe()함수는 데이터프레임객체에 바로 적용시키는 함수.

파라미터에 들어가는 매핑함수의 반환형태를 그대로 따라서 값을 반환

```python
# 타이타닉 데이터셋 불러와서 -> 'age', 'fare'열만 나오게
df = sns.load_dataset('titanic')[['age','fare']]
df
```

```python
# 값이 비어있는가에 대한 함수
def isNaN2(dataframe):
  return dataframe.isnull()
# 각 열의 비어있는 요소의 총 개수를 구해주는 함수
def countNaN2(dataframe):
  return isNaN2(dataframe).sum()
# 각 열의 비어있는 요소의 총 개수를 더한 값을 구해주는 함수
def total_number_isNaN2(dataframe):
  return countNaN2(dataframe).sum()
```

```python
# 순차적으로 적용
result_df = df.pipe(isNaN2)
result_df
```

```python
result_mc = df.pipe(countNaN2)
result_mc
```

```python
result_total_count = df.pipe(total_number_isNaN2)
result_total_count
```
