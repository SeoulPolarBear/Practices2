# 1. 시리즈 (Series)

Series : 객체를 담을 수 있는 1차원 배열스러운 자료구조

어떤 Numpy자료형도 가능 O


```python
# 1-1. 딕셔너리로 만들기 (dict)
# 시리즈는 딕셔너리 구조와 비슷함
dic = {'a':1,'b':2,'c':3,'d':4}

series_dic = pd.Series(dic)
print(type(series_dic))
series_dic
# 딕셔너리의 key값이 인덱스(index)로, value값이 값으로
```

```python
# 인덱스와 값은 아래처럼 확인
print(series_dic.index)
print(series_dic.values)
```

```python
# 직접 만들기도 O
# 1-2. 리스트로 만들기

obj = pd.Series([4,5,-4,-5])
obj
print(obj.index) # range(4)와 동일
print(obj.values)

# RangeIndex(start=0, stop=4, step=1)
# 샘플 데이터프레임을 만들 때 따로 인덱스를 입력하지 않았기 때문에 0부터 시작하는 인덱스가 부여(start = 0)
# 4개의 인덱스 (step=4) / 1씩 건너서 (step=1)
```

```python
obj2 = pd.Series([4,5,-5,-4], index=['a', 'b', 'c', 'd'])
obj2
```

```python
# 1-3. 튜플로 만들기
tup_data = ("Polarbear", " 2022-12-07", "남", True)
series_data = pd.Series(tup_data, index=["name", "date", "gender", "bool"])
series_data 
```

```python
# 1-4 . 요소선택, 인덱싱
print(series_data[0])
print(series_data["name"])
```

```python
#여러 값을 인덱싱 할 때 위에처럼 값만 반환하지 않고,
#짝을 이루는 인덱스와 값을 모두 반환

#888888888888888888888888888888888888주의할점
# 각각의 위치 or 이름으로 인덱싱할때는 대괄호를 2개를 씌워야함... [[]]
print(series_data[[1,2]])
print()
print(series_data[['date','bool']])
```

```python
# : 로 인덱싱 할 때는 대괄호 하나만 필요
print(series_data[1:3])
print()
print(series_data["date" : "gender"])
#인덱스의 이름으로 사용하면 범위 끝이 포함됨
```

```python
# Boolean배열을 사용해서 값 걸러내기 or 산술 연산 or 수학 함수 적용 O
obj2 = pd.Series([4,5,-5,-4], index=['a', 'b', 'c', 'd'])
print(obj2[obj2>0])
print()
print(obj2*2)
print()
print('b' in obj2)
print()
print('e' in obj2)
```

```python
ticket = {'busan' : 30000,'jeju' : 50000,'jeonju' : 40000,'dajeon' : 55000}
ticketS = pd.Series(ticket)
ticketS
```

```python
city = ['seoul','busan', 'jeju', 'jeonju']
a = pd.Series(ticket,index = city)
a # dajeon이 key로 없어서 안 들어간걸 알 수 있다,.
```

```python
a.isnull()
```

# 2. 데이터프레임 (DataFrame)

데이터프레임은 시리즈가 여러개 합쳐진 자료형

표, 엑셀과 같은 스프레드시트 형식의 자료구조

다양한 컬럼이 있고, 각각의 컬럼 안에는 서로 다른 종류의 

자료형(숫자, 문자열, 논리,...)이 담김

DataFrame : row(행), column(열)에 대한 index를 가지고 있음



```python
# 2-1. 딕셔너리로 만들기
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2' : [7,8,9], 'c3' : [10,11,12]}
# 1개의 열들이 1개의 시리즈, key : 열 이름 
df = pd.DataFrame(dict_data)

print(type(df))
print(df)
```

```python
# 2-2. 리스트로 만들기 1
#   같은 길이의 리스트에 담긴 dict or numpy를 이용

data = { 'time' : ['morning', 'afternoon', 'evening'],
          'startHour' : [6,12,18],
          'behavior' : ['getup', 'lunch', 'gohome']
          }
data1 = pd.DataFrame(data)
data1
```

```python
# 원하는 순서대로 columns을 지정하면, 그 순서대로 df 객체가 생성
pd.DataFrame(data, columns=['time', 'behavior','startHour'])
```

```python
# 2-2. 리스트로 만들기 2
df = pd.DataFrame([[21,'남자', '00대'], [20,'여자', 'ㅁㅁ대']],
                  index = ["Lim", "Kim"],
                  columns = ['age', 'gender', 'school'])

print(df)
print(df.index)
print(df.columns)
df
```

```python
# 2-3. 행, 열 이름 변경 : 기본적인 방식

df.index = ['student1', 'student2']
df.columns = ['나이','성별', '학교']
df

# 인덱스(index)와 열(colmns) 이름의 수에 맞게 지정해 주기!!
```

```python
# 2-4. 행, 열 이름 변경 : rename() 함수
# df.rename()
# df.rename(columns={변경전 열이름 : 변경 후 열이름, ...}, inplace = True)로 사용할 수 있고,
# inplace = True 옵션을 지정해주면 df에 바로 적용됨

df = pd.DataFrame([[21,'남자', '00대'], [20,'여자', 'ㅁㅁ대']],
                  index = ["Lim", "Kim"],
                  columns = ['age', 'gender', 'school'])
df
```

```python
# 열 이름 변경
df.rename(columns={'age' : '연령', 'gender' : '성별', 'school' : '소속'}, inplace=True)
df
```

```python
# 행 이름 변경 => index=로 지정
df.rename(index={'Lim' : 'Hwang', 'Kim' : 'Ahn'}, inplace=True)
df
```

# 행 삭제 : drop() 함수


```python
exam_data = {'math' : [90,80,70], 'korean' : [98,89,95], 'english' : [85,95,100], 'history' : [100,90,90]}
df = pd.DataFrame(exam_data, index=['student1','student2','student3'])
df
```

```python
# drop("행 or 열 이름", axis = 0 or 1,inplace = True)
# 행을 삭제하려면 axis = 0(기본값)
# 열을 삭제하려면 axis = 1 << 꼭 지정 !!!!

df2 = df.copy()

df2.drop('student1', inplace=True)
df2
```

```python
# 열 삭제
# 2개 이상의 열을 drop하고 싶으면 대괄호[] 안에 명시
df3 = df.copy()

df3.drop(['math', 'history'],axis = 1, inplace=True)
df3
```

```python
exam_data = {'math' : [90,80,70], 'korean' : [98,89,95], 'english' : [85,95,100], 'history' : [100,90,90]}
df = pd.DataFrame(exam_data, index=['kim','lee','park'])
df
```

```python
# kim과 lee의 점수 출력
print(df.iloc[:2])
print(df.loc[:'lee'])
# 두개 이상의 행을 선택할 때는 대괄호를 신경써야...!
# 각각 선택은 두개[[]] / 범위 (:)로 선택은 한개
l1= df.loc[['kim','lee']]
print(l1)
p1 = l1= df.iloc[[0,1]]
print(p1)
```

```python
#열 선택
# 열 선택은 따로 함수가 필요 x
# df. 열이름 or df['열이름']으로 선택이 가능
```

```python
# 수학 열만 선택
math1 = df['math']
print(math1)

print()

math2 = df.math
print(math2)
```

```python
# 범위 슬라이싱
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'], 
             'math' : [98,89,95,80,70,40,30], 
             'eng' : [100,80,70,85,100,60,100],
             'kor' : [98,90,90,70,60,50,60],
             'pt' : [98,95,90,100,40,70,80]}
df = pd.DataFrame(exam_data)
df
```

```python
# 모든 행을 2행 간격으로 출력
print(df.loc[::2])
```

```python
# 0행부터 5행까지 2행 간격으로 출력
print(df.iloc[0:6:2])
```

```python
# 전체 데이터 역순 정렬 > 모든 행을 -1행 간격
print(df.iloc[::-1])
print(df.loc[::-1])
```

```python
# 요소 선택
# set_index() 함수 : 특정한 열을 인덱스로 부여해주는 함수
```

```python
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'], 
             'math' : [98,89,95,80,70,40,30], 
             'eng' : [100,80,70,85,100,60,100],
             'kor' : [98,90,90,70,60,50,60],
             'pt' : [98,95,90,100,40,70,80]}
df = pd.DataFrame(exam_data)
df.set_index('name', inplace=True)
df
```

```python
# loc, iloc 행/열 선택과정에서
# 두 개 이상의 행이나 두 개 이상의 열을 선택하려면 [[]]을 이용
# ... 하나의 대괄호만 [] 쓴다면...?
df.loc['Ha','kor'] # 70 -> (ha,kor)에 있는 값 하나
```

```python
# 해당하는 요소 하나를 가져오라는 의미!
df.iloc[3,2]
```

```python
# Ah의 eng, kor 점수를 위에서 한 4가지 방법을 이용해서 출력
print(df.loc['Ah','eng':'kor'])
print()
print(df.iloc[2,1:3])
print()
print(df.loc['Ah',['eng','kor']])
print()
print(df.iloc[[2],[1,2]])
```

```python
# Oh, Woo의 kor,pt 점수를 4가지 방법으로 출력
print(df.loc['Oh':'Woo','kor':'pt'])
print()
print(df.iloc[0:2,1:3])
print()
print(df.loc[['Oh','Woo'],['kor','pt']])
print()
print(df.iloc[[0,1],[2,3]])
print(df.iloc[0:1,2:4])
```

```python
# 행, 열 추가
# 열 추가는 df['새로운 열이름'] = 값의 형태로 추가
# 새로운 열 이름이 아니라 기존 열 이름이라면... -> 덮어씌워진다.

exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'], 
             'math' : [98,89,95,80,70,40,30], 
             'eng' : [100,80,70,85,100,60,100],
             'kor' : [98,90,90,70,60,50,60],
             'pt' : [98,95,90,100,40,70,80]}
df = pd.DataFrame(exam_data)

df['music'] = 80
df
```

```python
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'], 
             'math' : [98,89,95,80,70,40,30], 
             'eng' : [100,80,70,85,100,60,100],
             'kor' : [98,90,90,70,60,50,60],
             'pt' : [98,95,90,100,40,70,80]}
df = pd.DataFrame(exam_data)
#요소 값 변경하기
df.loc[3] = 10 # 단일 값을 넣는다면 => 행 전체에는 모두 다 그  값으로 채워짐
df.loc[4] = ['Wa',100,100,100,100] # 행 또는 열의 수에 맞게 요소를 변경 / 추가
df.loc[7] = ['asdf',50,50,50,50] # 새로운 행 추가
df
```

```python
# 행과 열 바꾸는 : 전치행렬 transpose() 함수, T
exam_data = {'name' : ['Oh','Woo','Ah','Ha','Kim','Lee','Park'], 
             'math' : [98,89,95,80,70,40,30], 
             'eng' : [100,80,70,85,100,60,100],
             'kor' : [98,90,90,70,60,50,60],
             'pt' : [98,95,90,100,40,70,80]}
df = pd.DataFrame(exam_data)

print(df)
print()
df = df.transpose()
print(df)

df = df.T
print(df)
# 전치를 두번하면 원상태로 돌아옴!
```
