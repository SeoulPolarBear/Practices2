```python
#1. 남자아이, 여자아이 출생 수 평균을 구해주세요 (정수로 출력)
print(np.mean(birth[1], dtype=np.int64))
print(np.mean(birth[2], dtype=np.int64))
```

```python
#2. 2000년도부터의 데이터를 구해주세요 (2000 ~ 2011)
# 연도    남자    여자
# 2000    1234567 1234567
#...
data2000 = np.empty((0,3),np.int64)
for bir in birth:
  if bir[0] >= 2000 :
    data2000 = np.append(data2000, np.array([bir]), axis=0)
print(data2000)
#------------------------------------------------------------------------------------------
print("연도", "남자", "여자", sep="\t")
print("=" * 30)
for y, m, f in birth[birth[:,0] > 2000]:
  print(y,m,f,sep="\t")
```

```python
# 3. 남자아이 출생이 가장 많은 연도와 남자아이 출생 수를 구해주세요
max_boy = np.argmax(birth[:,1])
print(birth[max_boy,:1])
#------------------------------------------------------------------------------------------------------------
birth[birth[:,1].argmax()]
y = birth[birth[:,1].argmax()][0] #연도
b = birth[birth[:,1].argmax()][1] # 남자아이 출생 수

print("연도", "남자", sep="\t")
print(y, b, sep="\t")
```

```python
# 4. 여자아이 출생이 가장 많은 top5를 구하세요 (연도, 여자아이 수)
birth2 = np.array(birth)
print(birth2)
birth3 = np.array(birth[:,2])
print(birth3)
birth2_top5 = np.empty((0,3))
for _ in range(5):
  num1 = np.argmax(birth3)
  print(num1)
  birth2_top5 = np.append(birth2_top5, np.array([birth2[num1,:]]),axis=0)
  birth2 = np.delete(birth2, num1,axis=0)
  birth3 = np.delete(birth3, num1)
birth2_top5 = birth2_top5.astype('int64')
print(birth2_top5)


#------------------------------------------------------------------------------------------------------------
top5_girls = np.sort(birth[:,2])[-1: -6:-1] # 역순으로 찾을 때는 인덱스에 -1 끝에서부터 1칸씩
print("연도", "여자", sep="\t")
for i in birth:
  for j in top5_girls:
    if(i[2] == j):
      print(i[0], j, sep="\t")
```

```python
# 5. 여자아이 출생률(%)을 구하고, 각 연도별로 여자아이 출생수가 차지하는 비율을
#     데이터로 보여주세요. (소수점 둘째자리까지)
# 연도    남자    여자    여자출생률
#===========================================
# 1880    123123  1000000 0.41
birthWithPercent_girls = np.empty((0,3))
print("연도    남자    여자    여자출생률\n")
print("=============================================================\n")
for bir in birth:
  percent = np.round((bir[2] / (bir[1] + bir[2])), 2)
  # bir2 = np.append([bir],[percent])
  # print(bir2)
  #np.append(birthWithPercent_girls , np.array([bir]),axis=0)
  print("%d   %d    %d    %.2f" %(bir[0], bir[1], bir[2] ,percent))  
#   np.append(birthWithPercent_girls, bir2)
#   bir2 = np.array([])
# print(birthWithPercent_girls)

#---------------------------------------------------------------------------------------
girls_ratio = (birth[:,2] / (birth[:,1] + birth[:,2])).round(2)
print("연도","남자", "여자", sep="\t")
for y, b, g, gr in zip(birth[:,0],birth[:,1],birth[:,2],girls_ratio):
  print(y,b,g,gr, sep="\t")
```

```python
# 6. 2000년도 이후 데이터에 대해 출생량을 구하고, 
#     남자아이 수가 1800000 이상이면 '많음', 아니면 '적음' 이라는 내용을 출력
print(data2000)
for d in data2000:
  if d[1] >= 1800000:
    print(d[0] , "많음")
  else:
    print(d[0] , "적음")

#_--------------------------------------------------------------------------------------------
after_2000 = birth[birth[:,0] >= 2000]
boys = after_2000[:,1]
print(boys)

amount = np.where(boys >= 1800000, "많음", "적음")
for y, b, amt in zip(after_2000[:,0], boys, amount):
  print(y, b, amt)
```

```python
# 7. 1800년대, 1900년대, 2000년대의 각 남자아이, 여자아이 출생수의 평균을 구해주세요
data1800 = np.empty((0,3),np.int64)
for bir in birth:
  if bir[0] >= 1800 and  bir[0] < 1900:
    data1800 = np.append(data1800, np.array([bir]), axis=0)
print(data1800)

data1800_avg = np.mean(data1800, axis = 0)
print(data1800_avg)


data1900 = np.empty((0,3),np.int64)
for bir in birth:
  if bir[0] >= 1900 and bir[0] < 2000:
    data1900 = np.append(data1900, np.array([bir]), axis=0)
data1900_avg = np.mean(data1900, axis = 0)
print(data1900)
print(data1900_avg)

#-----------------------------------------------------------------------------------------------
print("1800년")
print("남자","여자")

k_1800s = birth[(birth[:,0] >= 1800) & (birth[:,0] < 1900)][:, 1:]
print(k_1800s.mean(axis=0)[0], k_1800s.mean(axis=0)[1])

print("1900년")
print("남자","여자")

k_1900s = birth[(birth[:,0] >= 1900) & (birth[:,0] < 2000)][:, 1:]
print(k_1900s.mean(axis=0)[0], k_1900s.mean(axis=0)[1])

print("2000년")
print("남자","여자")

k_1800s = birth[(birth[:,0] >= 2000)][:, 1:]
print(k_1800s.mean(axis=0)[0].round(2), k_1800s.mean(axis=0)[1])
```

```python
# 8. 여자아이 25% ~ 75% 사이의 값을 구해주세요 (연도, 여자아이 출생수)
percetages= np.quantile(birth[:,2],[0.25,0.75])
print(percetages)
girl_between = np.empty((0,2),np.int64)
for bir in birth:
  if bir[2] >= percetages[0] and  bir[2] < percetages[1]:
    # print(bir[0])
    # print(girl_between.ndim)
    # print(np.array([bir[0],bir[2]]).ndim)
    girl_between = np.append(girl_between, np.array([[bir[0],bir[2]]]), axis=0)
print(girl_between)

#-------------------------------------------------------------------------------------
g_q25 = np.quantile(birth[:,2],0.25)
g_q75 = np.quantile(birth[:,2],0.75)
print(birth[(birth[:,2] >= g_q25) & (birth[:,2] <= g_q75)][:,[0,2]])
```
